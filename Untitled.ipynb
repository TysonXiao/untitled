```python
def cz(idx,col,df):
    num = len(idx)
    for i in range(num-1):
        start = idx.iloc[i]
        end = idx.iloc[i+1]
        cz_num = end-start
        start_num = df.loc[start,col]
        end_num = df.loc[end,col]
        fillna = np.linspace(start_num,end_num,cz_num)
        df.loc[start:end,col] = fillna
    return df
```

# 计算到上一个非na时间的时间间隔

```python
df["last_notna_time"] = df[df["发动机转速"].notna()].Time   ##当时不是na的时间
```

```python
df_notna_time = df[df["发动机转速"].notna()].Time
```

```python
df_notna_time_diff = (df_notna_time - df_notna_time.shift(1)).dt.total_seconds()
```

```python
df["last_notna_time_diff"] = df_notna_time_diff
```

```python
df["last_notna_time_diff"] = df["last_notna_time_diff"].fillna(method = "bfill")
```

# 利用时间间隔分类处理缺失数据

```python
df.loc[df["last_notna_time_diff"]>180,"发动机转速"] = df.loc[df["last_notna_time_diff"]>180,"发动机转速"].fillna(0)  #长时间na的填充0
```

```python
df["发动机转速"] = df["发动机转速"].fillna(method = "ffill")
```
