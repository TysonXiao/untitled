```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
####读取数据
import os
```

```python
os.chdir(r"F:\sanyi\untitled-master\True")
true_pl = []
for f in os.listdir(r"F:\sanyi\untitled-master\True"):
    df = pd.read_csv(f)
    tmp = df["泵送排量"]
    
    max_ = tmp.max()
    min_ = tmp.min()
    median_ = tmp.median()
    mean_ = tmp.mean()
    
    tmp_diff = tmp - tmp.shift(1)
    l_30 = len(tmp_diff[tmp_diff>30])
    l_20 = len(tmp_diff[tmp_diff>20])
    l_n55 = len(tmp_diff[tmp_diff<-55])
    
    true_pl.append([max_,min_,median_,mean_,l_30,l_20,l_n55])
```

```python
df_t = pd.DataFrame(true_pl)
```

```python
df_t["label"] = 1
```

```python
os.chdir(r"F:\sanyi\untitled-master\False")
true_pl = []
for f in os.listdir(r"F:\sanyi\untitled-master\False"):
    df = pd.read_csv(f)
    tmp = df["泵送排量"]
    
    max_ = tmp.max()
    min_ = tmp.min()
    median_ = tmp.median()
    mean_ = tmp.mean()
    
    tmp_diff = tmp - tmp.shift(1)
    l_30 = len(tmp_diff[tmp_diff>30])
    l_20 = len(tmp_diff[tmp_diff>20])
    l_n55 = len(tmp_diff[tmp_diff<-55])
    
    true_pl.append([max_,min_,median_,mean_,l_30,l_20,l_n55])
df_f = pd.DataFrame(true_pl)
```

```python
df_f["label"] = 0
```

```python
df_f
```

```python
data = pd.concat([df_t,df_f])
```

```python
data.drop(["label"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
data.drop(["label"],axis=1), data["label"], test_size=0.33, random_state=42)
```

```python
 X_train, X_test, y_train, y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=4)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score
```

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
```

```python
# specify your configurations as a dict
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': {'l2', 'l1'},
    'num_leaves': 12,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

print('Starting training...')
# train
gbm = lgb.train(params,
                lgb_train,
                num_boost_round=20,
                valid_sets=lgb_eval,
                early_stopping_rounds=5)

print('Saving model...')
# save model to file
gbm.save_model('model.txt')

print('Starting predicting...')
# predict
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred = y_pred[0].apply(lambda x: 1 if x>0.5 else 0)
```

```python
# eval
print('The accuracy of prediction is:', accuracy_score(y_test, y_pred))
```
